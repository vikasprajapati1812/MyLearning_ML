# CAT Boost   (Categorical Boosting)

```python
# Data set---->Social_Network_Ads.csv
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
```

```python
data
```

```python
cat_feat=np.where(data.dtypes!=np.int64)[0]
```

```python
data.info()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
data.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.3)
catt=CatBoostClassifier(iterations=2000)
```

```python
catt.fit(xtrain,ytrain,cat_features=cat_feat,eval_set=(xtest,ytest))
```

```python
catt.predict(xtest)
```

```python
ypred=catt.predict(xtest)
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
confusion_matrix(ytest,ypred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(ytest,ypred)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=catt.classes_)
disp.plot()
```

```python
catt.score(xtest,ytest)
```

```python
catt.score(xtrain,ytrain)
```
