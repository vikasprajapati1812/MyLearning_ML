```python
def bub_sort(array): 
    for i in range(0, len(array)): 
        for j in range(0, len(array) - i - 1): 
            if array[j] < array[j + 1]: 
                temp = array[j] 
                array[j] = array[j + 1] 
                array[j+1] = temp 
    return array 
my_array = np.array([20,14,25,16,45,60,12,9])
```

```python
my_array
```

```python
print(bub_sort(my_array))
```
