```python
#splitting the data into X & Y
X=data.drop("Outcome",axis=1)
```

```python
Y=data.iloc[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
data.shape
```

```python
#Splitting the X & Y in tran and test 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,random_state=42,test_size=0.25,stratify=Y)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
#scaling data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
x_train=scale.fit_transform(xtrain)
x_test=scale.transform(xtest)
```

```python
#Modelling 
#Logistic 

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,ytrain)
```

```python
lr.predict(x_test)
```

```python
ypred=lr.predict(x_test)
```

```python
#performance check though metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(ypred,ytest)
```

```python
print(classification_report(ypred,ytest))
```

```python
confusion_matrix(ypred,ytest)
```

```python
#decison tree model
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,ytrain)
```

```python
dt.predict(x_test)
```

```python
ypreddt=dt.predict(x_test)
```

```python
accuracy_score(ypreddt,ytest)
```

```python
print(classification_report(ypreddt,ytest))
```

```python
confusion_matrix(ypreddt,ytest)
```

```python
#Random forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,ytrain)
```

```python
rf.predict(x_test)
```

```python
ypredrf=rf.predict(x_test)
```

```python
accuracy_score(ypredrf,ytest)
```

```python
print(classification_report(ypredrf,ytest))
```

```python
confusion_matrix(ypredrf,ytest)
```

```python
#xgboost
from xgboost import XGBRFClassifier
xg=XGBRFClassifier()
xg.fit(x_train,ytrain)
```

```python
ypredxg=xg.predict(x_test)
```

```python
accuracy_score(ypredxg,ytest)
```

```python
print(classification_report(ypredxg,ytest))
```

```python
confusion_matrix(ypredxg,ytest)
```

```python
#catboost
from catboost import CatBoostClassifier
ct=CatBoostClassifier()
ct.fit(x_train,ytrain)
```

```python
ct.predict(x_test)
```

```python
ypredct=ct.predict(x_test)
```

```python
accuracy_score(ypredct,ytest)
```

```python
print(classification_report(ypredct,ytest))
```

```python
confusion_matrix(ypredct,ytest)
```

```python
#Cross Validations  ----

from sklearn.model_selection import cross_val_score
print(cross_val_score(rf, x_train, ytrain, cv=5))
print(cross_val_score(lr, X, Y, cv=5))
```

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(rf, X, 
                        Y, )
sorted(scores.keys())
```

```python
#cross validation using rf model

scores = cross_validate(rf, X, Y,
                        scoring='precision_macro', cv=5,
                         return_estimator=True)
```

```python
scores
```

```python
#cross validation using lr(logistic Regression) model

scores = cross_validate(lr, X, Y,
                        scoring='precision_macro', cv=5,
                         return_estimator=True)
```

```python
scores
```

```python
#cross validation using dt model
scores = cross_validate(dt, X, Y,
                        scoring='precision_macro', cv=5,
                         return_estimator=True)
```
